```python
import pandas as pd
import numpy as np
import torch
device = "cuda"
```

```python
from mllms import QwenAddressParsingModel, ZeroShot, JsonDictPromptTemplate

prompt = "Parse this address, output a json dict with the keys \"City\" and \"StreetName\": %(examples)s\nAddress: %(address)s"


model = QwenAddressParsingModel(
    model_name="Qwen/Qwen3.5-9B",
    prompt=JsonDictPromptTemplate(prompt),
    example_strategy=ZeroShot(),
    device=device
)

model.parse_addresses(["123 Main St, Springfield, IL 62704"])
```
