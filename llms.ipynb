````python
import pandas as pd

PLACE_COLS = [
    "ApplicantBirthPlace", 
    "ApplicantCurrentAddress", 
    "VictimBirthPlace", 
    "VictimDeathPlace",
    "VictimCurrentAddress"
]

df = pd.read_json("data/1870.json")

print(f"Number of BZK cards: {len(df)}")

sampled = df.sample(n=10)
sampled.sample(n=10)
````

````python
from unittest import result
from transformers import pipeline
import json



PROMPTS = [
    "Segment the following address into components: {address}\n"
    "Output only a JSON object with the following keys: streetNumber, street, city, state, country, "
    "postcode, houseNumber, buildingName. "
    "You can add new keys but only if a component does not fit one of the given keys. "
    "Do not set any value that is not explicitly present in the address, "
    "fill fields that cannot be determined with null. "
    "Addresses may include german terms such as 'Straße' for street or its abbreviation 'Str.'. "
    "These may occur as a suffix to the street name.",

    "Annotate each component in the following address: {address}\n"
    "Consider the component types streetNumber, street, city, state, country, "
    "postcode, houseNumber, buildingName. "
    "You can add new types but only if a component does not fit one of the given types. "
    "Each component type should only occur once in each address. Not all addresses will contain all component types and you should not add the missing ones. "
    "Note that addresses may include german terms such as 'Straße' for street or its abbreviation 'Str.'. "
    "These may occur as a suffix to the street name. "
    "Output only a JSON list of [component, type] tuples. For example, the address \"Berlin, Alexanderplatz 1, 10178\" should be annotated as:\n"
    "```json\n[ [\"Berlin\", \"city\"], [\"Alexanderplatz\", \"street\"], [\"1\", \"streetNumber\"], [\"10178\", \"postcode\"] ]\n```"
    
]

class LlamaAddressSegmentationModel:
    def __init__(self, model_name, prompt, output_parser = None):
        self.pipe = pipeline("text-generation", model=model_name)
        self.prompt = prompt
        self.output_parser = output_parser or (lambda x : x)

    def segment_address(self, address: str) -> str:
        messages = [
            {"role": "user", "content": self.prompt.format(address=address)}
        ]
        result = self.pipe(messages)
        response = result[0]["generated_text"][1]["content"]
        return self.output_parser(response)

def extract_json_block(model_response : str):
    limit_chars = [('{', '}'), ('[', ']'), ('"', '"')]
    json_str = model_response
    parts = model_response.split("```")
    if len(parts) == 3:
        json_str = parts[1]
    elif len(parts) >= 3:
        for part in parts:
            part = part.strip()
            if part.startswith("json") or any(part.startswith(lim[0]) and part.endswith(lim[1]) for lim in limit_chars):
                json_str = part
                break
    if json_str.startswith("json"):
        json_str = json_str[4:].strip()
    return json_str

class JSONbjectParser:
    def __call__(self, response: str) -> dict:
        try:
            json_str = extract_json_block(response)
            obj = json.loads(json_str)
            for k in obj.keys():
                assert k not in ["fullResponse", "model-fullAddress", "parseError"], f"Key collision with reserved field: {k}"
                # It is plausible that the model just echoes back the full address in a field named "fullAddress"
                # But that would collide with our own field named "fullAddress" that we add later
                if k == "fullAddress":
                    obj["model-fullAddress"] = obj["fullAddress"] # avoid collision but keep the wrongfully generated field
            obj["fullResponse"] = response
            data = obj
        except json.JSONDecodeError as e:
            data = {"fullResponse": response, "parseError": f"JSONDecodeError: {e}"}
        return data

class JSONTuplesParser:
    def __call__(self, response: str) -> dict:
        try:
            json_str = extract_json_block(response)
            tuples = json.loads(json_str)
            data = {}
            for part, ptype in tuples:
                if isinstance(part, str) and part.strip() == "":
                    continue
                assert ptype not in ["fullResponse", "model-fullAddress", "parseError"], f"Key collision with reserved field: {ptype}"
                if ptype == "fullAddress":
                    data["model-fullAddress"] = data["fullAddress"] # avoid collision but keep the wrongfully generated field
                else:
                    collision_value = data.get(ptype)
                    if collision_value is None:
                        data[ptype] = part
                    else:
                        data[ptype] = collision_value + "___" + part
            data["fullResponse"] = response
        except json.JSONDecodeError as e:
            data = {"fullResponse": response, "parseError": f"JSONDecodeError: {e}"}
        return data

````

````python
from collections import OrderedDict
sample_addresses = df[PLACE_COLS].stack().sample(n=5, random_state=26012026)

def test_model(model, addresses=sample_addresses):
    # Ordered set of all keys seen in results (values don't matter, but there is no collection ordered set)
    col_names = OrderedDict((k, 0) for k in [
        "fullAddress", "fullResponse", "streetNumber", "street", "city", "state", "country", 
        "postcode", "houseNumber", "buildingName"]) 
    rows = []
    for (rid, field), address in addresses.items():
        print(f"Record ID: {rid} / Field: {field} / Address: {address}")
        response = model.segment_address(address)
        for k in response.keys(): 
            col_names[k] = 0  # add any new keys to the ordered set (again, value doesn't matter)
        response["fullAddress"] = address
        print(f"Parsed response: {response}")
        rows.append(response)
    return pd.DataFrame(rows, index=addresses.index, columns=col_names.keys())
        
````

````python
model = LlamaAddressSegmentationModel("meta-llama/Llama-3.2-1B-Instruct", PROMPTS[0], JSONbjectParser())
test_model(model).dropna(how="all", axis=1).fillna("")
````
