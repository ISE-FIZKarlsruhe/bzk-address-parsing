```python
import pandas as pd
import numpy as np

cols = pd.MultiIndex.from_product(
    [["A", "B"], ["x", "y"], ["value", "other"]],
    names=["lvl0", "lvl1", "lvl2"]
)

df = pd.DataFrame(
    np.random.randn(5, len(cols)),
    columns=cols
)

other_df = pd.Dataframe(["q", "w", "e", "r", "t"], columns=["foobar"])

df.loc[0, ("A", "x", "value")] = None
df.loc[1, ("B", "y", "value")] = None

df
```

```python
out = (
    df
    .xs("value", level=2, axis=1)     # keep only lowest-level column
    .stack([0, 1])                    # move lvl0, lvl1 into rows
    .reset_index()
)
out
```

```python
other_df = pd.DataFrame([[f"a{i}", f"b{i}"] for i in range(5)], columns=["A", "B"])
other_df
```

```python
s = out.merge(
        other_df
        .stack()
        .reset_index()
        .rename(columns={"level_1" : "lvl0", 0 : "testCol"}),
        on = ["level_0", "lvl0"],
        how="inner"
)
s
```
