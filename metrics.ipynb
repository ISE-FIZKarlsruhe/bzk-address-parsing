```python
import pandas as pd

def confusion_metrics(preds : pd.DataFrame, labels : pd.DataFrame):
    false_keys = [col for col in preds.columns if col not in labels.columns]
    # false positives means any span predicted that is not in the ground truth
    false_positives = preds[false_keys].notna().sum().sum()
    # false negatives means any span in the ground truth that is missing in the predictions
    false_negatives = 0
    # true positives means any span correctly predicted
    true_positives = 0
    for col in labels.columns:
        if col not in preds.columns:
            # all ground truth spans in this column are false negatives
            false_negatives += labels[col].notna().sum()
        else:
            matches = (labels[col] == preds[col])
            n_matches = matches.sum()
            true_positives += n_matches
            false_negatives += len(matches) - n_matches
    
    precision = true_positives / (true_positives + false_negatives)
    recall = true_positives / (true_positives + false_negatives)
    f1 = 2 * (precision * recall) / (precision + recall)
    accuracy = true_positives / (true_positives + false_positives + false_negatives)
    return {
        "recall" : recall,
        "precision" : precision,
        "f1" : f1,
        "accuracy" : accuracy
    }
```
